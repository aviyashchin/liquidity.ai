```python
df_h2o = h2o.import_file("theorem_output_factors.csv") 
```

```python
df_h2o[114] = df_h2o[114].asfactor()

df_h2o[41] = df_h2o[41].asfactor()
df_h2o[42] = df_h2o[42].asfactor()
df_h2o[44] = df_h2o[44].asfactor()
df_h2o[45] = df_h2o[45].asfactor()
df_h2o[62] = df_h2o[62].asfactor()
df_h2o[106] = df_h2o[106].asfactor()
```

```python
# Split into training, validation, and test sets
# Note that Column 0 is Listing Number
# Last column is Y
# Generate random numbers and create training, validation, testing splits 
r = df_h2o.runif()   # Random UNIForm numbers, one per row 
df_h2o_train_hex = df_h2o[r  < 0.7] 
df_h2o_valid_hex = df_h2o[(r >= 0.7) & (r < 0.8)]
df_h2o_test_hex = df_h2o[r >= 0.8] 
```

```python
df_h2o_train_hex.shape
```

```python
df_h2o_valid_hex.shape
```

```python
df_h2o_test_hex.shape
```

```python
# Features
X = df_h2o.columns[2:114]
```

```python
# Classification
Y = df_h2o.columns[114:115]
```

```python
Y
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
rf = H2ORandomForestEstimator(seed=12, ntrees=10, max_depth=20, balance_classes=False)
rf.train(x=X, y=Y, training_frame=df_h2o_train_hex, validation_frame=df_h2o_valid_hex)
```

```python
rf.show()
```

```python
pred = rf.predict(df_h2o_test_hex)
```

```python
pred.head()
```

```python
perf = rf.model_performance(df_h2o_test_hex)
```

```python
perf.show()
```

```python
def model(model_object, test):
        #predicting on test file
        pred = model_object.predict(test)
        pred.head()
        #Building confusion matrix for test set
        perf = model_object.model_performance(test)
        perf.show()
        print(perf.confusion_matrix())
        print(perf.precision())
        print(perf.accuracy())
        print(perf.auc())
```

```python
model(rf, df_h2o_test_hex)
```

### Deep Learning

```python
from h2o.estimators.deeplearning import H2ODeepLearningEstimator 
model = H2ODeepLearningEstimator(
    distribution="bernoulli",
    activation="RectifierWithDropout",
    hidden=[32,32,32],
    input_dropout_ratio=0.2,
    l1=1e-5,
    epochs=10, 
    balance_classes=True)
model.train(
    x=X,
    y=Y,
    training_frame=df_h2o_train_hex,
    validation_frame=df_h2o_valid_hex)
# The model runs for only 10 epochs since it is just meant as a trial run
```

```python
model.show()
```

```python
model_100 = H2ODeepLearningEstimator(
    distribution="bernoulli",
    activation="RectifierWithDropout",
    hidden=[32,32,32],
    input_dropout_ratio=0.2,
    l1=1e-5,
    epochs=100, 
    balance_classes=True)
model_100.train(
    x=X,
    y=Y,
    training_frame=df_h2o_train_hex,
    validation_frame=df_h2o_valid_hex)
```
